```python
import pandas as pd

df = pd.DataFrame({})
X_col_idxs = range(768)
dtypes = {idx: 'int8' for idx in X_col_idxs} #set dtypes beforehand to save memory
dtypes[768] = 'str'

df = pd.read_csv('./data/clean/2016_5k.csv', sep=',', header=None, dtype=dtypes, memory_map=True)
df.head()
```

```python
response = df.iloc[:,-1]
response = response.apply(lambda x : x[1:] if x.startswith("#") else x)

response = response.apply(lambda x: float(x[1:]) if x.startswith('+') 
                          else (-float(x[1:]) if x.startswith('-') else float(x)))

df.iloc[:,-1] = response
```
